### O que é o IMDb?

> **IMDb**, abreviação de **Internet Movie Database**, é uma base de dados online amplamente reconhecida e utilizada, que fornece informações sobre filmes, programas de televisão, elenco, equipes de produção, biografias de artistas e muito mais. Fundado em 1990, o IMDb se tornou uma das fontes mais abrangentes e confiáveis de informações relacionadas ao entretenimento. Ele permite que os usuários pesquisem e acessem uma vasta coleção de dados sobre filmes e programas de televisão, incluindo sinopses, classificações, resenhas de críticos e usuários, premiações e até mesmo informações triviais. O IMDb é frequentemente utilizado como referência por profissionais da indústria cinematográfica e por entusiastas de filmes e séries de todo o mundo.

### Dataset Utilizado

O dataset utilizado contém informações sobre o Top 1000 de séries e filmes do IMDB, que é uma lista classificatória dos filmes e séries mais bem avaliados pelos usuários do site. A classificação é determinada pela média ponderada das avaliações feitas pelos usuários registrados do IMDb.

A classificação é baseada em um algoritmo que leva em consideração vários fatores, incluindo o número de avaliações recebidas, a pontuação média das avaliações e a idade das avaliações. O algoritmo é projetado para dar maior peso às avaliações recentes, de modo que filmes mais antigos precisam de mais avaliações para obter uma classificação alta.

- Poster_Link - Link do pôster usado pelo IMDb
- Series_Title - Nome do filme
- Released_Year - Ano de lançamento do filme
- Certificate - Certificado obtido pelo filme
- Runtime - Tempo total de duração do filme
- Genre - Gênero do filme
- IMDB_Rating - Classificação do filme no site IMDb
- Overview - Breve história/resumo
- Meta_score - Pontuação obtida pelo filme
- Director - Nome do diretor
- Star1, Star2, Star3, Star4 - Nome das estrelas/atores
- No_of_votes - Número total de votos
- Gross - Valor arrecadado pelo filme

### Classificações indicativas

A coluna **certificate** possuí o análago às classificações indicativas que temos no Brasil.

No Brasil, as classificações indicativas são definidas pelo Ministério da Justiça, por meio do Departamento de Promoção da Diversidade Cultural (DPDC), e seguem uma estrutura diferente das classificações mencionadas no dataset. No entanto, é possível fazer um paralelo geral entre as classificações mencionadas e as classificações indicativas brasileiras:

- **U (Universal; suitable for every age):** Equivalente à classificação indicativa Livre (L), que indica que o conteúdo é adequado para todas as idades.
- **UA (Parental Guidance; is not always appropriate for kids under the age of 12):** Equivalente à classificação indicativa 10 anos (10), que indica que o conteúdo pode conter elementos que exigem o acompanhamento dos pais ou responsáveis para crianças menores de 10 anos.
- **A (Adult; not appropriate for children under the age of 18):** Equivalente à classificação indicativa 18 anos (18), que indica que o conteúdo é restrito a maiores de 18 anos, não sendo adequado para crianças ou adolescentes.
- **R (Restricted; contains some adult material; Under 17 requires accompanying parent or adult guardian):** Não possui uma equivalência direta nas classificações indicativas brasileiras. Poderia ser relacionado à classificação indicativa 16 anos (16), que indica que o conteúdo pode conter elementos impróprios para menores de 16 anos.
- **PG-13 (Parents Strongly Cautioned; Some material may be inappropriate for children under 13):** Equivalente à classificação indicativa 14 anos (14), que indica que o conteúdo pode conter elementos que exigem o acompanhamento dos pais ou responsáveis para crianças menores de 14 anos.
- **PG (Parental Guidance Suggested; Some material may not be suitable for children. Parents urged to give "parental guidance"):** Equivalente à classificação indicativa 12 anos (12), que indica que o conteúdo pode conter elementos que exigem o acompanhamento dos pais ou responsáveis para crianças menores de 12 anos.
- **G (General Audience; All ages admitted):** Equivalente à classificação indicativa Livre (L), que indica que o conteúdo é adequado para todas as idades.

Para definir se duas features estão correlacionadas, será utilizado a correlação de Pearson.

>A correlação de Pearson é uma medida estatística que avalia a relação linear entre duas variáveis contínuas. Ela mede a força e a direção dessa relação, variando entre -1 e 1.
O coeficiente de correlação de Pearson é calculado utilizando a seguinte fórmula:
>
>$$
r = \frac{\sum((X - \bar{X}) \cdot (Y - \bar{Y}))}{\sqrt{\sum(X - \bar{X})^2} \cdot \sqrt{\sum(Y - \bar{Y})^2}}
>$$
>O valor de r indica a intensidade e a direção da correlação:
>- Se r for próximo de 1, indica uma correlação positiva forte, o que significa que as variáveis tendem a aumentar juntas.
>- Se r for próximo de -1, indica uma correlação negativa forte, o que significa que as variáveis tendem a ter uma relação >-inversa: quando uma variável aumenta, a outra diminui.
>- Se r for próximo de 0, indica uma correlação fraca ou inexistente entre as variáveis.



Avaliaremos também o modelo de regressão criado utilizando coeficiente de determinação, também conhecido como R²

> Coeficiente de determinação é uma medida estatística que quantifica a proporção da variabilidade total da variável dependente que é explicada pelo modelo de regressão.
> - Um valor de R² próximo de 0 indica que o modelo não explica a variação da variável dependente e não há relação entre as variáveis.
> - Um valor de R² próximo de 1 indica que o modelo explica a maior parte da variação da variável dependente e há uma relação forte entre as variáveis.
>
>Ele é definido como:
>$$
R² = 1 - (SSE / SST)
>$$
> Onde
>- SSE (Sum of Squared Errors) é a soma dos quadrados dos resíduos ou erros entre os valores previstos pelo modelo e os valores reais da variável dependente.
> - SST (Total Sum of Squares) é a soma dos quadrados das diferenças entre os valores reais da variável dependente e a média desses valores.

```python
def getR2(df, col1, col2):
	model = LinearRegression()
	X = df[[col1]]
	y = df[col2]
	model.fit(X, y)
	y_pred = model.predict(X)
	return r2_score(y, y_pred)
```

#### Correalação entre duração do filme e o rating

```python
np.corrcoef(x=df_imdb['run_time'], y=df_imdb['imdb_rat'])
```

Na matriz resultante do comando np.corrcoef, que calcula o coeficiente de correlação entre duas variáveis, temos na posição (0, 0) da matriz a correlação entre a variável x (runtime) e ela mesma, na posição (1, 1), temos a correlação entre a variável y (imdb_rat) e ela mesma, ou seja, a diagonal principal da matriz representa as correlações das variáveis consigo mesmas.

Os valores fora da diagonal principal representam a correlação entre as variáveis x e y. Como o valor obtido é mais próximo de 0 do que de 1, observamos uma correlação fraca entre a duração do filme e seu rating no IMDB.

```python
sea.regplot(x='run_time', y='imdb_rat', data = df_imdb).set_title('Regressão Linear: IMBD Rating e Duração do Filme')
```

```python
getR2(df_imdb, 'run_time', 'imdb_rat')
```

Observamos que, como as duas variáveis tem uma fraca correlação, o modelo de refressão linear apresetou um coeficiente de correlação mais próximo de 0.
